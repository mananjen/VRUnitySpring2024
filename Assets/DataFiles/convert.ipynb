```python
import pandas as pd


file_path = 'E:\\Spring 2024 VR\\omicron-unity\\Assets\\DataFiles\\athyg_31_reduced_m10.csv'


df = pd.read_csv(file_path)

df.columns
```

```python

selected_columns = df[['id', 'hip', 'dist', 'x0', 'y0', 'z0', 'absmag', 'mag', 'vx', 'vy', 'vz', 'spect']].copy()


new_file_path = 'selected_columns_athyg_31_reduced_m10.csv'

```

```python
selected_columns.dropna(subset=['x0', 'y0', 'z0', 'spect'], inplace=True)
selected_columns['spect'] = selected_columns['spect'].str[0]
selected_columns = selected_columns[selected_columns['spect'].isin(['O', 'B', 'A', 'F', 'G', 'K', 'M'])]
```

```python
selected_columns.to_csv(new_file_path, index=False)
```

```python
selected_columns.head()
```

```python
KM_SEC_TO_PC_YEAR = 1.0220122e-6

selected_columns["vx_pc_year"] = selected_columns["vx"] * KM_SEC_TO_PC_YEAR
selected_columns["vy_pc_year"] = selected_columns["vy"] * KM_SEC_TO_PC_YEAR
selected_columns["vz_pc_year"] = selected_columns["vz"] * KM_SEC_TO_PC_YEAR

converted_velocities = selected_columns[['id', 'hip', 'dist', 'x0', 'y0', 'z0', 'absmag', 'mag', 'vx_pc_year', 'vy_pc_year', 'vz_pc_year', 'spect']].copy()

new_file_path = 'selected_with_velocities.csv'
```

```python
converted_velocities.to_csv(new_file_path, index=False)
```
